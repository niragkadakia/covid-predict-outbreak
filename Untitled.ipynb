```python
import warnings
warnings.filterwarnings('ignore')

import numpy as np
import csv
import matplotlib.pyplot as plt
from datetime import datetime, date, timedelta
import urllib, json, requests
from geopy.geocoders import Nominatim
from scipy.stats import linregress, pearsonr
```

```python
# Sets timestamp for 1-1-2020 to 0
INIT_DATE = 18262
COVID_DATA_FILE = '../covid-19-data/us-counties.csv'
TRANSIT_DATA_FILE = '../transit/January 2020 Raw Database.txt'
POPULATION_DENSITY_FILE = '../pop-density/DEC_10_SF1_GCTPH1.US05PR_with_ann.csv'
MIN_REGRESS_PTS = 20
MAX_POP_DENSITY = 1000
MAX_POP = 1e6
MIN_NUM_CASES = 50
```

### Load the COVID data by county

```python
# Get all unique counties by fips
counties = []
with open(COVID_DATA_FILE, newline='') as csvfile:
    reader = csv.reader(csvfile, delimiter=',', quotechar='|')
    for iR, row in enumerate(reader):
        if iR == 0:
            fips = 0
        elif row[3] == '':
            if (row[1]) == 'New York City':
                fips = 36061
            else:
                fips = 0
        else:
            fips = int(row[3])
        counties.append(fips)

# All counties in same order as data        
counties = np.array(counties)
county_idxs = np.unique(counties)
num_counties = len(county_idxs)
```

```python
covid_data = [None for i in range(num_counties)]
num_cases = np.zeros(num_counties)
with open(COVID_DATA_FILE, newline='') as csvfile:
    reader = csv.reader(csvfile, delimiter=',', quotechar='|')
    for iR, row in enumerate(reader):
        if iR == 0:
            continue
        
        date_str = row[0]
        yr = int(date_str.split('-')[0])
        mh = int(date_str.split('-')[1])
        dy = int(date_str.split('-')[2])
        dt = datetime(yr, mh, dy)
        elap = int(round(dt.timestamp())/3600/24) - INIT_DATE
        
        county = counties[iR]
        county_idx = np.where(county_idxs == county)[0][0]
        
        cases = int(row[4])
        deaths = int(row[5])
        
        if covid_data[county_idx] is None:
            covid_data[county_idx] = np.ndarray((3, 1))
            covid_data[county_idx][:, 0] = [elap, cases, deaths]
        else:
            covid_data[county_idx] = np.vstack((covid_data[county_idx].T, np.array([elap, cases, deaths]))).T
for iR in range(num_counties):
    num_cases[iR] = np.cumsum(covid_data[iR][1, :])[-1]
```

### Fit each counties case growth rate to exponential growth rate

```python
inf_rates = np.zeros(num_counties)
for idx in range(num_counties):
    x = covid_data[idx][0]
    y = np.log(np.cumsum(covid_data[idx][1]))/np.log(2)
    nonzero = np.where(np.isfinite(y))[0]
    if len(nonzero) == 0:
        inf_rates[idx] = np.nan
        continue
    x_nonzero = x[nonzero[0]:]
    y_nonzero = y[nonzero[0]:]
    if len(x_nonzero) < MIN_REGRESS_PTS:
        inf_rates[idx] = np.nan
        continue
    nominal_slopes = []
    nominal_rerrs = []
    for cut in range(MIN_REGRESS_PTS//2):
        #ls, yint = np.polyfit(x_nonzero[cut:], y_nonzero[cut:], 1)
        ls, yint, rerr, perr, pval = linregress(x_nonzero[cut:], y_nonzero[cut:])
        nominal_slopes.append(ls)
        nominal_rerrs.append(rerr)
    inf_rates[idx] = nominal_slopes[np.argmax(nominal_rerrs)]

fig = plt.figure(figsize=(12, 3))
plt.subplot(121)
inf_rates = np.array(inf_rates)
plt.hist(1./inf_rates[np.isfinite(inf_rates)], bins=50, color='dodgerblue', normed=True)
plt.xlabel('Doubling times')
plt.ylabel('Pdf')
plt.axvline(1./inf_rates[county_idxs == 36061], color='r')
plt.axvline(1./inf_rates[county_idxs == 36119], color='g')

plt.subplot(122)
inf_rates = np.array(inf_rates)
plt.hist(inf_rates[np.isfinite(inf_rates)], bins=30, color='dodgerblue', normed=True)
plt.xlabel('Growth rates (1/day)')
plt.ylabel('Pdf')
plt.axvline(inf_rates[county_idxs == 36061], color='r')
plt.axvline(inf_rates[county_idxs == 36119], color='g')
plt.show()
```

### Get population density statistics

```python
pop_stats = np.empty((num_counties, 3))
pop_stats[:] = np.nan
with open(POPULATION_DENSITY_FILE, newline='') as csvfile:
    reader = csv.reader(csvfile, delimiter=',', quotechar='|')
    for iR, row in enumerate(reader):
        if iR < 3:
            continue
        county_idx = np.where(county_idxs == int(row[4]))[0]
        if len(county_idx) > 0:
            try:
                pop = int(row[7])
            except ValueError:
                pop = int(row[7].split('(')[0]) 
            pop_stats[county_idx[0], :] = [pop, float(row[12]), float(row[13])]
plt.figure(figsize=(12, 3))
plt.subplot(121)
plt.hist(pop_stats[np.isfinite(pop_stats[:, 0]), 0]/1e6, bins=np.arange(0, 2, 0.03), color='dodgerblue')
plt.yscale('log')
plt.xlabel('Population in millions')
plt.subplot(122)
plt.hist(pop_stats[np.isfinite(pop_stats[:, 2]), 2]/1e3, bins=np.arange(0, 3, 0.1), color='dodgerblue')
plt.yscale('log')
plt.xlabel('Population density in 1000s per sq. mi.')
plt.show()
```

### Infection growth rate versus household population density

```python
ok_idxs = np.isfinite(pop_stats[:, 2])
ok_idxs *= np.isfinite(inf_rates)
ok_idxs *= pop_stats[:, 2] <= MAX_POP_DENSITY
ok_idxs *= num_cases > MIN_NUM_CASES

# Best fit line
_sl, _yint, _, _, _ = linregress(pop_stats[ok_idxs, 2], inf_rates[ok_idxs])
_xs = np.arange(MAX_POP_DENSITY)
_ys = _xs*_sl + _yint
plt.plot(_xs, _ys, color='r')
plt.xlabel('Population density (pop/sq. mi.)')
plt.ylabel('Infection growth rate base 2')
plt.scatter(pop_stats[ok_idxs, 2], inf_rates[ok_idxs], alpha=0.5, color='dodgerblue')
plt.show()

print ('corr, pval = ', pearsonr(pop_stats[ok_idxs, 2], inf_rates[ok_idxs]))
print ('growth rate increase per 1000 pop_dens increase', _sl*1e3)
```

### Infection growth rate versus population

```python
ok_idxs = np.isfinite(pop_stats[:, 0])
ok_idxs *= np.isfinite(inf_rates)
ok_idxs *= pop_stats[:, 0] <= MAX_POP
ok_idxs *= num_cases > MIN_NUM_CASES

# Best fit line
_sl, _yint, _, _, _ = linregress(pop_stats[ok_idxs, 0], inf_rates[ok_idxs])
_xs = np.arange(MAX_POP)
_ys = _xs*_sl + _yint
plt.plot(_xs/1e6, _ys, color='r')
plt.xlabel('Population in millions')
plt.ylabel('Infection growth rate base 2')

plt.scatter(pop_stats[ok_idxs, 0]/1e6, inf_rates[ok_idxs], alpha=0.5, color='dodgerblue')
plt.show()

print ('corr, pval = ', pearsonr(pop_stats[ok_idxs, 0], inf_rates[ok_idxs]))
print ('growth rate increase per 1 million people', _sl*1e6) 
```
